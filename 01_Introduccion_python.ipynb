### 2.1 Markdown

Markdown es un lenguaje de marcas que se utiliza comúnmente para formatear texto en documentos en línea. Fue creado originalmente por John Gruber y Aaron Swartz en 2004 y se ha convertido en una de las opciones más populares para escribir y formatear contenido en la web debido a su simplicidad y facilidad de uso.

Markdown se utiliza para crear documentos de texto enriquecido que incluyen formato de texto, encabezados, listas, enlaces, imágenes y otros elementos comunes de la web, todo ello utilizando una sintaxis de marcado sencilla y legible por humanos. El objetivo principal de Markdown es permitir a los escritores concentrarse en el contenido del texto en lugar de preocuparse por la complejidad del formato.

Algunas de las características más comunes de Markdown incluyen:

Encabezados: Puedes crear encabezados de diferentes niveles utilizando el símbolo #.

Listas: Puedes crear listas numeradas y viñetas con facilidad.

Enlaces: Puedes crear enlaces a otras páginas web.

Imágenes: Puedes incrustar imágenes en tus documentos.

Texto en negrita y cursiva: Puedes aplicar formato de negrita y cursiva a partes del texto.

Markdown se utiliza en una variedad de contextos, incluyendo la creación de documentos técnicos, blogs, documentación de software, correos electrónicos y más. La simplicidad de Markdown y su capacidad de ser convertido a HTML u otros formatos hace que sea una elección popular para escribir contenido web y colaborativo en plataformas como GitHub, Stack Overflow y, entre muchas otras, para escribir texto explicativo dentro de un cuaderno de Jupyter.

#### 3.4.2 `while`

Un bucle `while` especifica una condición y un bloque de código que se ejecuta hasta que la condición se evalua a `False` o se termina el bucle con `break`.

````python
x = 256

total = 0

while x > 0:
  if total > 500:
    break
  total += x
  x = x // 2
````

#### 3.4.3 `for`

Se puede avanzar a la siguiente iteración de un bucle utilizando la *keyword* `continue`.

````python
sequence = [1, 2, None, 4, 5]

total = 0

for value in sequence:
  if value is None:
    continue
  total += value
````

````python
total
````

Un bucle puede abortarse utilizando la *keyword* `break`.

````python
sequence = [1, 2, 0, 4, 6, 5 , 2, 1]

total_until_5 = 0

for value in sequence:
  if value == 5:
    break
  total_until_5 += value
````

````python
total_until_5
````

## 6 Compresiones de listas, conjuntos y cadenas

Las comprensiones de listas permiten formar de forma concisa una nueva lista filtrando los elementos de una colección y transformando los elementos que pasan el filtro en una expresión concisa.

`[expr for val in collection if condition] `

Equivale a

```
result = []
for val in collection:
  if condition:
    result.append(expr)
```

Para diccionarios y conjuntos de usa la siguientes expresiones:  

`dict_comp = {key_expr : value_expr for value in collection if condition}`

`set_comp = {expr for value in collection if condition}`


````python
lyrics = ['come', 'as', 'you', 'are',',', 'as', 'you', 'were']
````

````python
[cadena for cadena in lyrics]
````

````python
[len(cadena) for cadena in lyrics]
````

````python
{cadena: len(cadena) for cadena in lyrics}
````

````python
{len(cadena) for cadena in lyrics}
````
